# WHERE (조건절)

사용자가 자신이 원하는 자료만을 검색하기 위해서 사용하며 FROM 절 다음에 위치한다.

#### 조건식의 구성
- 칼럼명(보통 조건식의 좌측)
- 비교 연산자
- 문자, 숫자, 표현식 (보통 조건식의 우측)
- 비교 칼럼명(JOIN 사용시)

## 연산자의 종류

WHERE 절에 사용되는 연산자는 3가지 종류가 있다.

### ✅ 비교 연산자(부정 비교 연산자 포함)
- = : 같다
- '>' : 보다 크다 
- < : 보다 작다
- '>=' : 보다 크거나 같다
- <= : 보다 작거나 같다
- != : 같지 않다
- ^= : 같지 않다
- <> : 같지 않다(ISO 표준으로 모든 운영체제에서 사용 가능한 연산자.)
- NOT COLUMN_NAME = : ~와 같지않다
- NOT COLUMN_NAME > : ~보다 크지않다

숫자 유형 칼럼의 경우 숫자로 변환이 가능한 문자열과 빅되면 상대 타입을 숫자 타입으로 바꾸어 비교한다.

→ WHERE HEIGHT >= '170' 이라고 표현하더라도 HEIGHT 칼럼이 숫자 유형의 변수이므로 내부적으로 '170' 문자열을 숫자인 170으로 바꾸어 처리함.

### ✅ SQL 연산자(부정 SQL 연산자 포함)
- BETWEEN A AND B : A와 B 값 사이에 있으면 TRUE ( A와 B의 값이 포함되는 이상/이하 의 개념 )
- NOT BETWEEN A AND B : A와 B값 사이에 없으면 TRUE 
- IN LIST : 리스트에 있는 값 중에서 어느 하나라도 일치하면 TRUE
- NOT IN LIST : 리스트에 있는 값 중에서 어느 하라라도 일지하지 않으면 TRUE
- LIKE 비교문자열 : 비교문자열과 형태가 일치하면 된다. (%, _ 사용)
- IS NULL : NULL값인 경우 TRUE
- IS NOT NULL : NULL값이 아닌 경우 TRUE
---
#### 활용 TIP_1

다중 리스트를 이용한 IN 연산자 활용 가능

WHERE (JOB,DEPTNO) IN (('MANAGER',20),('CLERK',30));

→ JOB이 MANAGER이면서 DEPTNO가 20인 경우 OR JOB이 CLERK이면서 DEPTNO가 30인경우

!= WHERE JOB IN ('MANAGER','CLERK') AND DEPTNO IN (20,30); 

→ 위의 조합과 동일한 결과는 물론 MANAGER와 30, CLERK와 20의 조합도 출력됨

---
#### 활용 TIP_2

와일드 카드를 사용한 LIKE 연산자 활용
- % : 0개 이상의 어떤 문자를 의미함
- _ : 1개의 단일 문자를 의미함

EX) WHERE PLAYER_NAME LIKE '장%';   →   PLAYER_NAME 칼럼에서 '장' 으로 시작하는 문자열을 모두 출력

---
#### 활용 TIP_3

NULL값? → 값이 존재하지 않는 것으로 확정되지 않은값을 표현함. 공백이나  0과 달리 비교 자체가 불가능한 값!

NULL 값의 연산
- NULL값과의 수치연산은 NULL값을 리턴한다.
- NULL값과의 비교연산은 FALSE를 리턴한다.
- 어떤 값과 비교할 수도 없으며, 특정 값보다 크거나 작다고 표현할 수 없다.

---

### ✅ 논리 연산자
- @ AND @ : 앞에 있는 조건과 뒤에 오는 조건이 TRUE이면 결과도 TRUE (동시만족)
- @ OR @ : 앞에 있는 조건과 뒤에 오는 조건 중 어느 하나라도 TRUE이면 결과도 TRUE
- NOT @ : 뒤에 오는 조건에 반대되는 결과를 되돌려 준다.



### 연산 우선순위
괄호 → NOT연산자 → 비교연산자, SQL 연산자 → AND → OR 

연산 우선순위를 고려하지 못한 잘못된 예!

1. WHERE TEAM_ID = 'K02' OR TEAM_ID = 'K03'

2. AND POSITION = 'MF'

3. AND HEIGHT >= 170

4. AND HEIGHT <= 180;

WHAT'S WRONG? : 1번부터 4번까지 4가지 조건을 중첩하여 만족하는 ROW를 출력하고 싶었으나, OR보다 AND 조건이 우선 연산되어 버림

1. WHERE (TEAM_ID = 'K02' OR TEAM_ID = 'K03')

2. AND POSITION = 'MF'

3. AND HEIGHT >= 170

4. AND HEIGHT <= 180;

원래 의도에 맞는 결과를 얻기 위해서는 괄호를 적절히 사용하여야 한다.



## ROWNUM(ORACLE)  / TOP(SQL SERVER)

### ROWNUM

각 행에 임시로 부여하는 일련번호인 ROWNUM은 원하는 만큼의 행을 가져오고 싶을 때 행의 개수를 제한하는 목적으로 사용한다.

SELECT PLAYER_NAME 

FROM PLAYER 

WHERE ROWNUM = 숫자 ;
